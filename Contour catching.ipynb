```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pyautogui as pe
```

```python
image = cv2.imread("C:/Users/hh492146/OneDrive - GSK/Desktop/b8.png")
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
copy_img = image.copy()
plt.imshow(image)
```

```python
_, binary = cv2.threshold(gray, 235, 255, cv2.THRESH_BINARY_INV)

# show it
plt.imshow(binary, cmap="gray")
plt.show()
```

```python
# find the contours from the thresholded image
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

print('Number of contours = ' + str(len(contours)))
print(contours[0])
```

```python
# draw all contours
image = cv2.drawContours(image, contours, -1, (0, 255, 3), 3)
```

```python
# show the image with the drawn contours
plt.imshow(image)

plt.savefig('C:/Users/hh492146/OneDrive - GSK/Desktop/contoured.png')
```

```python
c = max (contours,key = cv2.contourArea)
```

```python
c
```

```python
x,y,w,h= cv2.boundingRect(c)
```

```python
cv2.rectangle(image, (x,y),(x+w,y+h),(0, 255, 0),5)
plt.imshow(image)
```

```python
cropped_image = copy_img[y:y+h, x:x+w]
```
